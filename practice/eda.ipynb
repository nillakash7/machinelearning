### Plotting for Exploratory data analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load the dataset
pd.options.display.max_columns = None
# To make this notebook's output identical for every run
np.random.seed(0)
data = pd.read_csv('../data/practice/iris.csv')
# Display data randomly
data.head()
```

### Dimension of the data

```python
# Display the number of rows and columns in the dataset
print(data.shape)
```

### List of columns

```python
print(data.columns)
```

### How many records for each category/dependent variable are present

```python
data["variety"].value_counts()
```

### 2-D scatter plot

```python
data.plot(kind='scatter', x='sepal.length', y='sepal.width')
plt.title('Sepal Length vs Sepal Width')
```

```python
data.plot(kind='scatter', x='petal.length', y='petal.width')
plt.title('Petal Length vs Petal Width')
```

### 2-D scatter plot with colored data points

```python
sns.set_style("whitegrid")
sns.FacetGrid(data, hue="variety", height=5).map(plt.scatter, "sepal.length", "sepal.width").add_legend()
sns.FacetGrid(data, hue="variety", height=5).map(plt.scatter, "petal.length", "petal.width").add_legend()
plt.show()
```

### Observations

Setosa can be linearly separable from Versicolor and Virginica<br>
If a new data point falls in the range sepal.length (4.5 to 6.5) & sepal.width (2.8 to 4.5) as well as petal.length (0-3) & petal.width (0-1), it can be categorized as Setosa<br>
Separating Versicolor from Virginica is challenging using sepal.length & sepal.width but can be linearly separable using petal.length (3-5) & petal.width (1 - 1.75)
